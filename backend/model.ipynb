```python
import pandas as pd
```

```python
# Load the dataset
disease_symptoms_df = pd.read_csv('./csvData/disease_symptoms.csv', sep=",")
print("Disease Symptoms")
print(disease_symptoms_df.head())
```

```python
print("Disease Medicine")
disease_medicine_df = pd.read_csv('./csvData/disease_medicine2_utf8only.csv', sep=",")
print(disease_medicine_df.head())
```

```python
print("Disease Precaution")
disease_precaution_df = pd.read_csv('./csvData/disease_precaution.csv', sep=",")
print(disease_precaution_df.head())
```

```python
print("Disease Risk Factors")
disease_risk_factors_df = pd.read_csv('./csvData/disease_riskFactors2_utfonly.csv', sep=",")
print(disease_risk_factors_df.head())
```

```python
# Set indexes
disease_symptoms_df.set_index('Disease', inplace=True)
disease_risk_factors_df.set_index('DNAME', inplace=True)
print(disease_symptoms_df.head())
```

```python
disease_precaution_df.set_index('Disease', inplace=True)
```

```python
disease_medicine_df.set_index('Disease_ID', inplace=True)
```

```python
df2 = disease_symptoms_df.copy()
df3 = disease_precaution_df.copy()
df4 = disease_risk_factors_df.copy()
df5 = disease_medicine_df.copy()

df6 = df2.join(df3, how='outer')
df7 = df6.join(df4, how='outer')
df8 = df7.join(df5, how='outer', on='DID') # All tables joined

print(df8.head())
```

```python
df8.reset_index(inplace=True)
print(df8.head())
```

```python
print(df8.to_json())
```

```python
df10 = disease_symptoms_df.copy()
for i in df10.columns[0:]:
    df10[i] = df10[i].astype('category')
```

```python
cat_columns = df10.select_dtypes(['category']).columns
df10[cat_columns] = df10[cat_columns].apply(lambda x: x.cat.codes)
print(df10.head())
```

```python
df11 = disease_symptoms_df.copy()
df11 = df11.reset_index()
df11 = df11.melt(id_vars=['Disease'], var_name='Symptom', value_name='Symptom_Value')
print(df11.head())
```

```python
df12 = df11.drop(columns=['Symptom'])
print(df12.head())
```

```python
df13 = df12.pivot_table(index='Disease', columns='Symptom_Value', aggfunc='size', fill_value=0)
print(df13.head())
```
